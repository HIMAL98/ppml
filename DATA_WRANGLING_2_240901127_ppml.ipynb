```python
import pandas as pd
data={'Name':['Sai','Yuvasri','Poovannan','Sabharish','Sangamithra','Rithika','Rubasri'],'Age':[20,18,19,20,18,17,19],'Gender':['M','F','M','M','F','F','F'],'Marks':[90,76,'NaN',74,65,'NaN',71]}
df=pd.DataFrame(data)
print("Original DataFrame:\n",df)
c=avg=0
for ele in df['Marks']:
  if str(ele).isnumeric():
    c+=1
    avg+=ele
avg/=c
df=df.replace(to_replace="NaN",value=avg)
print("\nReplacing NaN with average marks:\n",df)

df=df[df['Marks']>=75]
df=df.drop(['Age'],axis=1)
print("\nFiltering and Dropping func:\n",df)

one=pd.DataFrame({'Name':['Sai','Sub','Poo','Yuvashri','Rithika'],'subject_id':['sub2','sub4','sub3','sub6','sub5'],'Marks_scored':[86,95,87,99,91]},index=[1,2,3,4,5])
two=pd.DataFrame({'Name':['Bhavana','Gopi','Sai'],'subject_id':['sub1','sub7','sub2'],'Marks_scored':[75,80,90]},index=[6,7,8]) # Added definition for 'two'
print("\nOriginal DataFrame 1:\n",one)
print("\nOriginal DataFrame 2:\n",two)
print(pd.concat([one,two]))

```
